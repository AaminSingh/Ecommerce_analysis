```python
#drop the rows and columns with missing values
df.dropna(subset = ["Description"], inplace = True)
```

```python
#is there data is missing customer id, we suppose is Guest transaction
df_with_customer = df.dropna(subset= ["CustomerID"]).copy()
df_with_customer["CustomerID"] = df_with_customer["CustomerID"].astype("int")
```

```python
#check Iscancelled Payment and Invoice number and check Data type "confirm false",'cancelled true'
df['isCancelled'] = df['InvoiceNo'].astype(str).str.startswith("C")
df['isCancelled']
```

```python
~df['isCancelled']
```

```python
#Create a new DataFrame only keeping  Valid Transaction
#Keep only those rows where the order is not cancelled, the quantity is greater than 0, and the unit price is greater than 0
df_clean = df[
   
    (~df['isCancelled']) &           #Keep only rows there iscancelled is false
    (df['Quantity']>0)  &            #keep only those rows whose quantity is greater than 0  Because a negative quantity usually represents a returned or cancelled item.
    (df['UnitPrice']>0)              #Keep only and check rows Unitprice is greater than 0
    ].copy()                         #create a separate copy of the Filtered DataFrame       

#Without .copy(), changes made to df_clean may also affect the original DataFrame or produce a SettingWithCopyWarning.

#With .copy(), df_clean is a separate DataFrame that you can safely modify.      
```

```python
df_clean
```

```python
df_clean['InvoiceDate'] = pd.to_datetime(df_clean["InvoiceDate"]) #"Convert this text into actual date and time values After conversion, Pandas understands that these are dates.
# Now the column's data type changes from:
# object (string)
# to
# datetime64[ns]
#Once the column is in datetime format, you can perform date-related operations.
#The converted datetime values replace the original string values in the DataFrame.
df_clean["InvoiceDate"]
# InvoiceDate Column
#         │
#         ▼
# Stored as Text (String)
#         │
#         ▼
# pd.to_datetime()
#         │
#         ▼
# Converted into Date & Time Format
#         │
#         ▼
# Stored back into InvoiceDate column
#         │
#         ▼
# Display the InvoiceDate column
```

```python
df_clean["TotalPrice"] = np.multiply(df_clean["Quantity"],df_clean["UnitPrice"])
#The purpose of this code is to calculate the total price of each product purchased.
#It tells us how many items were purchased.
#It tells us the price of one item.
#np.multiply() is a NumPy function that multiplies two values.
#It works like the * operator but can multiply entire columns element by element.

df_clean
```

```python
print("STEP3: Data Cleaned")
print("Original rows:",df.shape[0],"|Clean rows:",df_clean.shape[0])
print("Cancelled order removed:",df["isCancelled"].sum())
print(df_clean[["InvoiceNo","Quantity","UnitPrice","TotalPrice","InvoiceDate"]])
print("\n")
```

```python
#extract the year from the 'InvoiceDate' column and store it in a new column called 'Year'
df_clean['Year'] = df_clean['InvoiceDate'].dt.year
#extract the month (1-12) from the 'invoiceDate' column
df_clean['Month'] = df_clean['InvoiceDate'].dt.month
#Extract the day of the month (1-31) from the 'InvoiceDate' column
df_clean['Day'] = df_clean['InvoiceDate'].dt.day
#extract the hour (0-23) from the 'InvoiceDate' column and store it in a new column called 'Hour'
df_clean['Hour'] = df_clean['InvoiceDate'].dt.hour

#extract the day of the week (0-6) from the 'InvoiceDate' column and store it in a new column called 'DayOfWeek'
df_clean['DayOfWeek'] = df_clean['InvoiceDate'].dt.dayofweek
```
