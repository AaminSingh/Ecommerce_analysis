```python
#drop the rows and columns with missing values
df.dropna(subset = ["Description"], inplace = True)
```

```python
#is there data is missing customer id, we suppose is Guest transaction
df_with_customer = df.dropna(subset= ["CustomerID"]).copy()
df_with_customer["CustomerID"] = df_with_customer["CustomerID"].astype("int")
```

```python
#check Iscancelled Payment and Invoice number and check Data type "confirm false",'cancelled true'
df['isCancelled'] = df['InvoiceNo'].astype(str).str.startswith("C")
df['isCancelled']
```

```python
~df['isCancelled']
```

```python
#Create a new DataFrame only keeping  Valid Transaction
#Keep only those rows where the order is not cancelled, the quantity is greater than 0, and the unit price is greater than 0
df_clean = df[
   
    (~df['isCancelled']) &           #Keep only rows there iscancelled is false
    (df['Quantity']>0)  &            #keep only those rows whose quantity is greater than 0  Because a negative quantity usually represents a returned or cancelled item.
    (df['UnitPrice']>0)              #Keep only and check rows Unitprice is greater than 0
    ].copy()                         #create a separate copy of the Filtered DataFrame       

#Without .copy(), changes made to df_clean may also affect the original DataFrame or produce a SettingWithCopyWarning.

#With .copy(), df_clean is a separate DataFrame that you can safely modify.      
```

```python
df_clean
```

```python
df_clean['InvoiceDate'] = pd.to_datetime(df_clean["InvoiceDate"]) #"Convert this text into actual date and time values After conversion, Pandas understands that these are dates.
# Now the column's data type changes from:
# object (string)
# to
# datetime64[ns]
#Once the column is in datetime format, you can perform date-related operations.
#The converted datetime values replace the original string values in the DataFrame.
df_clean["InvoiceDate"]
# InvoiceDate Column
#         │
#         ▼
# Stored as Text (String)
#         │
#         ▼
# pd.to_datetime()
#         │
#         ▼
# Converted into Date & Time Format
#         │
#         ▼
# Stored back into InvoiceDate column
#         │
#         ▼
# Display the InvoiceDate column
```

```python
df_clean["TotalPrice"] = np.multiply(df_clean["Quantity"],df_clean["UnitPrice"])
#The purpose of this code is to calculate the total price of each product purchased.
#It tells us how many items were purchased.
#It tells us the price of one item.
#np.multiply() is a NumPy function that multiplies two values.
#It works like the * operator but can multiply entire columns element by element.

df_clean
```

```python
df_clean['TotalPrice']
```

```python
print("STEP3: Data Cleaned")
print("Original rows:",df.shape[0],"|Clean rows:",df_clean.shape[0])
print("Cancelled order removed:",df["isCancelled"].sum())
print(df_clean[["InvoiceNo","Quantity","UnitPrice","TotalPrice","InvoiceDate"]])
print("\n")
```

```python
#extract the year from the 'InvoiceDate' column and store it in a new column called 'Year'
df_clean['Year'] = df_clean['InvoiceDate'].dt.year
#extract the month (1-12) from the 'invoiceDate' column
df_clean['Month'] = df_clean['InvoiceDate'].dt.month
#Extract the day of the month (1-31) from the 'InvoiceDate' column
df_clean['Day'] = df_clean['InvoiceDate'].dt.day
#extract the hour (0-23) from the 'InvoiceDate' column and store it in a new column called 'Hour'
df_clean['Hour'] = df_clean['InvoiceDate'].dt.hour
#extract the day of the week (0-6) from the 'InvoiceDate' column and store it in a new column called 'DayOfWeek'
df_clean['DayOfWeek'] = df_clean['InvoiceDate'].dt.dayofweek
```

```python
print("STEP4: Date features added")
print(df_clean[["InvoiceDate","Year","Month","DayOfWeek","Hour"]].head(10))
print("\n" )
```

```python
#step5"Key Business Metrics
#Calulate the total revenue generated from all valid transactions.
#Formula:Total Revenue = Sum of all TotalPrice values

total_revenue = np.sum(df_clean['TotalPrice'])

#Calculate the total number of unique orders(invoices)
#nunique() counts unique InvoiceNo values
#If an invoice contains multiple products, it is still counted as ONE order
total_orders = df_clean['InvoiceNo'].nunique()

#Calculate the total numbers of unique customers.
#nunique()  counts distinct CustomerID values 
#If a customer places multiple orders,they are counted only once.
total_customers = df_clean["CustomerID"].nunique()

#Calculate the Average Order Value(AOV)
#Step1:Group all rows by InvoiceNo (each invoice may contain multiple products)
#step2:Sum the TotaLPrice of all products within each invoice.
#step3: Calculate the average of all invoice totals
#(Calculates the average amount spent per order by first finding the total value of each invoice and then averaging those totals.)
avg_order_value = df_clean.groupby("InvoiceNo")["TotalPrice"].sum().mean()


print(total_revenue)
print(total_orders)
print(total_customers)
print(avg_order_value)


# Original Data

# InvoiceNo    TotalPrice
# 1001            20
# 1001            30
# 1002            40
# 1002            10
# 1003            50
#         │
#         ▼
# groupby("InvoiceNo")
#         │
#         ▼
# 1001 → 20, 30
# 1002 → 40, 10
# 1003 → 50
#         │
#         ▼
# sum()
#         │
#         ▼
# 1001 → 50
# 1002 → 50
# 1003 → 50
#         │
#         ▼
# mean()
#         │
#         ▼
# Average Order Value = 50
```

```python
#Groups the data by product name, calculates the total revenue earned from each product, 
# sorts the products from highest to lowest revenue, and returns the top 10 products.
top_products_revenue = df_clean.groupby("Description")["TotalPrice"].sum().sort_values(ascending=False).head(10)
#Groups the data by product name, calculates the total quantity sold for each product, s
# orts the products from highest to lowest quantity sold, and returns the top 10 products.
top_products_quantity = df_clean.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)
print("Step6:Top 10 Prodcuts by Revenue")
print(top_products_revenue)
print("Step6:Top 10 Prodcuts by Quantity")
print(top_products_quantity)
```

```python
#STEP 7:Country-wise Analysis
country_revenue = (
    df_clean.groupby("Country")["TotalPrice"]
    .sum()
    .astype(int)
    .sort_values(ascending=False)
)
print("Step7:Revenue by Country")
print(country_revenue.head(10))

# Original DataFrame
#         │
#         ▼
# Group rows by Country
#         │
#         ▼
# Select TotalPrice column
#         │
#         ▼
# Add all sales of each country
#         │
#         ▼
# Convert revenue to integers
#         │
#         ▼
# Sort from highest revenue to lowest
#         │
#         ▼
# Store in country_revenue
```

```python
#STEP8: MONTHLY SALES TREND
#create  a new column containing only the year and month from the invoicedate
df_clean["YearMonth"] = df_clean["InvoiceDate"].dt.to_period("M")

#Calculate the total sales(revenue) for each Year-month 
monthly_sales = df_clean.groupby("YearMonth")["TotalPrice"].sum()
print("Step8:Monthly Sales Trend")
print(monthly_sales)
```

```python
df_clean["WeekMonth"] = df_clean["InvoiceDate"].dt.to_period("W")
weekly_sales = df_clean.groupby("WeekMonth")["TotalPrice"].sum()
print("Step8:Weekly Sales Trend")
print(weekly_sales)
```

```python
df_clean["DayWeek"] = df_clean["InvoiceDate"].dt.to_period("D")
daily_sales = df_clean.groupby("DayWeek")["TotalPrice"].sum()
print("Step8:Daily Sales Trend")
print(daily_sales)
```

```python
#STEP9:CUSTOMER ANALYSIS(RFM-style basics)
#createa customer summary by calculating spending,orders,average order value, and last purchase date, then sort customers by highest total spending 

customer_summary = df_clean.groupby("CustomerID").agg(
    TotalSpent = ("TotalPrice","sum"),
    NumOrders =("InvoiceNo","nunique"),
    AvgOrderValue = ("TotalPrice","mean"),
    LastPurchase = ("InvoiceDate","max")
).sort_values("TotalSpent",ascending=False)

customer_summary.head()
```

```python
#STEP10:NUMPY BASED STATISTICAL ANALYSIS ON ORDER VALUES

#calculate the total value of each order (invoice) and store the results as a Numpy array
order_values = df_clean.groupby("InvoiceNo")["TotalPrice"].sum().values
order_values
```

```python
#step 10:CALCULATE AND DISPLAY BASIC STATISTICS FOR ORDER VALUES
print("Step 10: Order Value Statistics(Numpy )")
#Display the average order value 
print("Mean",np.mean(order_values))

#Display the middle order value
print("Median",np.median(order_values))

#display the smallest order value
print("Minimum",np.min(order_values))

#display the spread of order values
print("Standard deviation",np.std(order_values))

#display the largest order value
print("Maximum",np.max(order_values))
```

```python
#step11: SAVE CLEANED DATA AND SUMMARY OUTPUTS

#save the cleaned dataset to a CSV file without including the row index
df_clean.to_csv("Cleaned _ecommerce_data.csv",index = False)

#save the customer summay report to a CSV file 
customer_summary.to_csv("customer_summary.csv")

#save the country-wise revenue report to a CSV file
country_revenue.to_csv("country_revenue.csv")
```
